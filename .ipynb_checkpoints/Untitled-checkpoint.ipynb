```python
import numpy as np
import re
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from matplotlib import cm
from datetime import datetime
import glob
import os
import json
import pickle
import six
sns.set()
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.options.mode.chained_assignment = None
import chardet
import warnings
warnings.filterwarnings("ignore")
```

```python
#Instantiate path with the directory of where the files are stored
path = r"C:\Users\user\Desktop\Data Science\Data Analytics\Data"
#assign the file format you are interested in to extension
extension = 'csv'
#Change the file directory to the path
os.chdir(path)
#Using the for loop iterate through the path searching for files with the format extension
all_csv = [i for i in glob.glob('*.{}'.format(extension))]
#print the list of files matching the files format found in the directory
all_csv
```

```python
#Itereste through the files assigned to all_csv, read and append the data to data_frames
# list to store each data frame separately
all_dataframes = [] 
for csv in all_csv:
    df = pd.read_csv(csv, encoding = 'ISO-8859-1')
    #Create a country column and add the respective country abbreviation to the column for eac data set
    df['country'] = csv[0:2] 
    all_dataframes.append(df)

#index 0 to 9 for [CA, DE, FR, GB, IN, JP, KR, MX, RU, US] datasets
all_dataframes[0].head()
```

<font size = +2>Correcting the data types for each column</font>

```python
#Using for loop iterate through the files in the all_dataframes
for df in all_dataframes:
    #Video ID Column
    df['video_id'] = df['video_id'].astype('str')
    
    #Trending Date Column
    df['trending_date'] = df['trending_date'].astype('str') #Assign string attribute to our Trending Date column
    #Split the values in the column using '.' as the delimiter
    date_pieces = (df['trending_date'].str.split('.'))
    #Create three columns, one each for Year, Month and Day.
    #Convert the values to Int
    df['Year'] = date_pieces.str[0].astype(int)
    df['Month'] = date_pieces.str[2].astype(int)
    df['Day'] = date_pieces.str[1].astype(int)
    
    #correcting the year in each column to 4 digits by adding 2000 to each value in the year column
    updatedyear = []
    for i in range(len(df)):
        y = df.loc[i, "Year"]
        newdate = y + 2000
        updatedyear.append(newdate)
    
    #Replace the values in the column 'Year' with the values in updatedyear[]
    for x in range(len(df)):
        z = updatedyear[x]
        #Get the value in the respective row of the df['Year'] column
        xy = df.loc[x, "Year"]
        #Replace the value in the column with the value from updatedyear[]
        df['Year'].replace(to_replace = xy, value = z, inplace=True)
        
        
    #Delete the columns 'trending_date', and assign the merged values from the Year, month and day column
    del df['trending_date']
    df['trending_dates'] = pd.to_datetime(df[['Year', 'Month', 'Day']], format = "%Y-%m-%d")
    
    #Delete the Year, Month and Day column they no longer needed
    del df['Year']
    del df['Month']
    del df['Day']
    
    #Title Column
    df['title'] = df['title'].astype('str')
    
    #channel_title Column
    df['channel_title'] = df['channel_title'].astype('str')
    
    #category_id Column
    df['category_id'] = df['category_id'].astype(str) 
    
    #tags Column
    df['tags'] = df['tags'].astype('str')
    
    #The Columns views, likes, dislikes, comment_count are already in correct data types i.e int64
    
    #thumbnail_link Column
    df['thumbnail_link'] = df['thumbnail_link'].astype('str') 
    
    #description Column
    df['description'] = df['description'].astype('str')
    
    #comments_disabled Column
    df['comments_disabled'] = df['comments_disabled'].astype('category') 
    
    #ratings_disabled Column
    df['ratings_disabled'] = df['ratings_disabled'].astype('category') 
    
    #Video_error_or_removed Column
    df['video_error_or_removed'] = df['video_error_or_removed'].astype('category') 
    
    #Convert the column publish_time to date format
    df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce', format='%Y-%m-%dT%H:%M:%S.%fZ')
        
```

```python
#Separating publish_time into date and time
for df in all_dataframes:
    #Inserting a new column for the publish date
    df.insert(4, 'publish_date', df['publish_time'].dt.date)
    #Convert the date and time format in the publish_time column to time instead
    df['publish_time'] = df['publish_time'].dt.time

#Formatting the publish_date to Year-Month-day
for df in all_dataframes:
    df['publish_date'] = pd.to_datetime(df['publish_date'], format = "%Y-%m-%d")
```

<font size = +1> Ratio of likes to dislike for each category </font>

```python
#group the datasets and sum up the aggregate likes and dislikes
likes = combined_data.groupby('category')['likes'].agg('sum')
dislikes = combined_data.groupby('category')['dislikes'].agg('sum')

#find the ratio of likes to dislikes
ratio_ld = likes/dislikes
ratio_ld = ratio_ld.sort_values(ascending=False).reset_index()
```

```python
#Assigning column names to our dataset
ratio_ld.columns = ['category', 'ratio_likes']
plt.subplots(figsize=(10, 15))
sns.barplot(x="ratio_likes", y="category", data=ratio_ld,
            label="Likes-Dislikes Ratio", color="b")
```

The ratio of likes to dislikes is calculated to reflect the proportion of likes to dislikes, this ratio was then assigned to the different categories present in the data using a Bar chart. According to the Bar chart, Pet & Animals has the highest ratio of likes to dislikes, depicting that this category has less dislikes compared to other categories. Shows and 'News & Politics has the least ratio of likes to dislikes.
